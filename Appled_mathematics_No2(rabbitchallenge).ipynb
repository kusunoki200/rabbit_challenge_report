## 第一章 線形代数

- スカラー・・・いわゆる普通の数（四則演算が可能）
- ベクトル・・・スカラーを任意の数並べて組にしたもの、「大きさ」、「向き」を持つ
- 行列・・・スカラーを表のように並べたもの、ベクトルを並べたもの

### [ ベクトル】
$n$個の実数$x_1,x_2,...,x_n$について、それらを縦に並べたものを縦ベクトル、横に並べたものを横ベクトルという。

$$
 \boldsymbol{x} =
     \left(\begin{array}{c}
     x_1 \\
     x_2 \\
     \vdots \\
     x_n
     \end{array}\right) \quad
 \boldsymbol{x} = (x_1,x_2,\dots,x_n)
$$

基本的には実数で考えるが、数以外のものが入ることもある。
ベクトルと言われたときに縦なのか横なのかは文脈で判断するが、縦で書かれていることが多い。縦ベクトルは転置すると横ベクトルになる。逆も成り立つ

### 【行列】
$m \times n$個の$x_{11},x_{12},\dots,x_{1n}, \dots, x_{m1},\dots,x_{mn}$をm行n列に並べたものを行列という。

$$
 \boldsymbol{X} = \left(
 \begin{array}{cccc}
 x_{11} & x_{12} & \dots & x_{1n}\\
 x_{21} & x_{22} & \dots & x_{2n}\\
 \vdots & \vdots &  & \vdots & \\
 x_{m1} & x_{m2} & \dots & x_{mn}\\
 \end{array}
 \right)
$$

行列は縦ベクトルを横に並べたものと解釈することもできる。

### 【行列とベクトルの積】
ベクトルに行列をかけることでベクトルを変換することができる。
- 例題

$$
\left(
\begin{array}{cc}
6 & 4 \\
3 & 5
\end{array}
\right)
\left(
\begin{array}{c}
1 \\
2
\end{array}
\right)
=
\left(
\begin{array}{cc}
6 \times 1 + 4 \times 2 \\
3 \times 1 + 5 \times 2
\end{array}
\right)
=
\left(
\begin{array}{c}
14 \\
13
\end{array}
\right)
$$

ベクトル$(1\ \ 2)^T$が$(14\ \ 13)^T$に変換されている

$$



### 【単位行列と逆行列】
行列$\bf{A}$について、$\bf{A}$にある行列$\bf{I}$をかけたとき、$\bf{A}$のままになるような行列$\bf{I}$を単位行列といい、下記のように表される。
$$
\bf{I} = 
\left(
\begin{array}{cc}
1 & 0 \\
0 & 1
\end{array}
\right)
$$

また、行列$\bf{A}$について、$\bf{A}$にある行列をかけたとき、その結果が単位行列になるような行列を逆行列といい$\bf{A}^{-1}$と表す。
$$
\bf{A}^{-1}\bf{A} = \bf{A}\bf{A}^{-1} = \bf{I} 
$$

$$
\left(

\right) \\

### 【逆行列の存在しない条件】
連立方程式において、解がない・解が１つに決まらない方程式がある。
$$
\left\{
\begin{eqnarray}
x_1+4x_2&=&7 \\
x_1+4x_2&=&10
\end{eqnarray}
\right. \quad
$$

$$
\left\{
\begin{eqnarray}
x_1+4x_2&=&7 \\
2x_1+8x_2&=&14
\end{eqnarray}
\right.
$$

以上のような連立方程式の係数を抜き出した行列は逆行列をもたない.

$$
\left(
\begin{array}{cc}
1 & 4\\
1 & 4\\
\end{array}
\right) \quad
\left(
\begin{array}{cc}
1 & 4\\
2 & 8\\
\end{array}
\right) \\
$$

$2 \times 2$の行列
$$
A = 
\left(
\begin{array}{cc}
a & b\\
c & d\\
\end{array}
\right) \\
$$

において、
$$
ad - bc = 0
$$
となるとき、$2 \times 2$の行列$\bf{A}$は逆行列を持たない。
このときの$ad-bc$を行列式といい、$det{A}$と表記する。
行列式は正方行列のみ計算できる。

### 【行列式の計算方法（3次以上）】
3次以上の行列についてはdetAは余因子展開で計算される。
- 余因子展開　\
$n$次の正方行列に$A$について第i行と第j列を取り除いた行列を$A_{ij}$とするとき
$$
\tilde{A}_{ij} = (-1)^{i+j} detA_{ij}
$$
$\tilde{A}_{ij}$を行列Aの$(i,j)$成分の余因子という

- n次正方行列の行列式の余因子展開

n次正方行列は余因子を用いて、
$$
det{A} = \sum_{j=1}^{n} \tilde{A}_{ij} a_{ij} = \sum_{i=1}^{n} \tilde{A}_{ij} a_{ij}
$$
と書くことができる。


- 3次正方行列で計算
$$
A =
\left(
\begin{array}{ccc}
a & b & c\\
d & e & f\\
g & h & i
\end{array}
\right) \\
$$
について、余因子は
$$
\tilde{A}_{11}=a\times det
\left(
\begin{array}{cc}
e & f\\
h & i
\end{array}
\right) \\
\tilde{A}_{21}=-d\times det
\left(
\begin{array}{cc}
b & c\\
h & i
\end{array}
\right) \\
\tilde{A}_{31}=g\times det
\left(
\begin{array}{cc}
b & c\\
e & f
\end{array}
\right) \\
$$
と書けるため、行列式は

$$
 det A = a\times det
\left(
\begin{array}{cc}
e & f\\
h & i
\end{array}
\right) 
-d\times det
\left(
\begin{array}{cc}
b & c\\
h & i
\end{array}
\right)
+g\times det
\left(
\begin{array}{cc}
b & c\\
e & f
\end{array}
\right)
$$

で計算できる。4次以上については手順を繰り返す。

### 【余因子展開による逆行列の求め方】

$$
 A^{-1} = \frac{1}{det A}A^{*}
$$

$(i, j)$成分の余因子を$\tilde{A}_{ij}$としたときに、

$$
 A^{*} =
\left(
\begin{array}{ccc}
\tilde{A}_{11} & \dots & \tilde{A}_{n1}\\
\vdots & & \vdots \\
\tilde{A}_{1n} & \dots & \tilde{A}_{nn}
\end{array}
\right)
$$
と書ける。$A^{*}$の(i,j)成分が$\tilde{A}_{ji}$と逆転しているのに注意

- 例
$$
\begin{array}{|ccc|}
1 & 0 & -1 \\
3 & 1 & 0 \\
2 & -1 & 1
\end{array}
=
1
\begin{array}{|cc|}
1 & 0  \\
-1 & 1
\end{array}
-3
\begin{array}{|cc|}
0 & -1  \\
-1 & 1
\end{array}
+2
\begin{array}{|cc|}
0 & -1  \\
-1 & 0
\end{array}
= 6
$$

###  【固有値と固有ベクトル】

ある行列$A$をベクトル$\boldsymbol{x}$にかけたとき、次のような式が成立する$\boldsymbol{x} \neq 0$のベクトルを固有ベクトルという。

$$
A \boldsymbol{x} = \lambda \boldsymbol{x}
$$

このとき$\lambda$はスカラーで、行列$A$をベクトルにかけることで元のベクトルの$\lambda$倍となり、向きは変わらず、大きさだけが変わる。$\lambda$を固有値と呼ぶ。

- 具体例
$$
\left(
\begin{array}{cc}
1 & 4 \\
2 & 3
\end{array}
\right)
\left(
\begin{array}{c}
1 \\
1
\end{array}
\right)
=
5
\left(
\begin{array}{c}
1 \\
1
\end{array}
\right)
$$

という関係があるとき
固有値$\lambda=5$、固有ベクトル$\boldsymbol{x} = (1\ 1)^T$となる

### 【固有値と固有ベクトルの求め化】

$$
 A \boldsymbol{x} = \lambda \boldsymbol{x} \\
 (A -\lambda I)\boldsymbol{x} = 0
$$

$x \neq 0$であることから、$A-\lambda I$の逆行列が存在しないことが条件となる。⇒$det A = 0$を満たす$\lambda$が固有値となる。

行列
$$
A =
\left(
\begin{array}{cc}
1 & 4 \\
2 & 3
\end{array}
\right)
$$
について、
$$
 |A -\lambda I| = 0 \\
 \begin{array}{|cc|}
 1- \lambda & 4 \\
 2 & 3-\lambda
 \end{array}
 = 0 \\
 (1-\lambda)(3-\lambda) - 4\cdot2 = 0 \\
 \lambda = 5 \ {or}\  -1
$$

よって行列Aの固有値は5と-1となる。それぞれの固有値に対応する固有ベクトルは、
- $\lambda = 5$
$$
\left(
\begin{array}{cc}
1 & 4\\
2 & 3
\end{array}
\right)
\left(
\begin{array}{c}
x_1 \\
x_2
\end{array}
\right)
=
5
\left(
\begin{array}{c}
x_1 \\
x_2
\end{array}
\right)
$$

$$
x_1 = x_2
$$

- $\lambda = -1$
$$
\left(
\begin{array}{cc}
1 & 4\\
2 & 3
\end{array}
\right)
\left(
\begin{array}{c}
x_1 \\
x_2
\end{array}
\right)
=
-1
\left(
\begin{array}{c}
x_1 \\
x_2
\end{array}
\right)
$$

$$
x_1 = -2x_2
$$


以上より、
- $\lambda = 5$のとき$\boldsymbol{x} = (1\ 1)^T$の定数倍が固有ベクトル
- $\lambda = -1$のとき$\boldsymbol{x} = (2\ -1)^T$の定数倍が固有ベクトル 

実際のところ固有ベクトルは条件を満たせばどれでもいいが、ベクトルの大きさを１に正規化したもの示すことも多い。
また、$detA = 0$の方程式を固有方程式と呼ぶ。

### 行列の固有値分解

$n$次正方行列$A$について、固有値$\lambda_1, \lambda_2, \dots, \lambda_n$、対応する固有ベクトルを$\boldsymbol{v}_1, \boldsymbol{v}_2, \dots, \boldsymbol{v}_n$とする。

このとき、固有値を対角成分に持つ行列$\Lambda$、固有ベクトルを列成分にもつ行列$V$について、
$$
 A = V \Lambda V^{-1}
$$
と変形することができる。
ここで、
$$
 \Lambda = 
 \left(
 \begin{array}{cccc}
 \lambda_1 & & &\\
  & \lambda_2 & &\\
  & & \ddots & \\
  & & & \lambda_n
 \end{array} 
 \right)
$$

$$
 V = (\boldsymbol{v}_1, \boldsymbol{v}_2, \dots, \boldsymbol{v}_n)
$$
としている。このように変換することで、行列$A$の累乗計算が容易になる。

- 例

行列
$$
A =
\left(
\begin{array}{cc}
1 & 4 \\
2 & 3 
\end{array}
\right)
$$

の固有値分解する。
固有値は、
$$
\begin{eqnarray*}
 \begin{array}{|cc|}
 1 - \lambda & 4 \\
 2 & 3 - \lambda
 \end{array}
 &= 0 \\
 (\lambda +1)(\lambda -5) &= 0 \\
 \lambda &= 5 , -1
\end{eqnarray*}
$$

$\lambda=5$の固有ベクトルは$(1\ 1)^T$, $\lambda=-1)$の固有ベクトルは$(1\  -\frac{1}{2})^T$となるので、
$$
 A =
 \left(
 \begin{array}{cc}
 1 & 1 \\
 1 & -1/2
 \end{array}
 \right)
 \left(
 \begin{array}{cc}
 5 & 0 \\
 0 & -1
 \end{array}
 \right)
 \left(
 \begin{array}{cc}
 1 & 1 \\
 1 & -1/2
 \end{array}
 \right)^{-1}
$$
と分解することができる。

#### 【対角化の条件】

$n$次正方行列$A$について、

①固有値がすべて異なるとき対角化可能　\
②Aに一時独立な固有ベクトルがn個存在すれば対角化可能また、逆も成り立つ

#### 【対称行列の固有値分解について】

考える行列が対称行列の場合、次が成り立つ

①固有値はすべて実数 \
②固有ベクトルは互いに直交するようにとることができる。

#### 【特異値分解】

正方行列でない$m \times n$の行列$M$を考える
この行列について
$$
\begin{eqnarray*}
M\boldsymbol{v} &= \sigma \boldsymbol{u} \\
M^T \boldsymbol{u} & = \sigma \boldsymbol{v}
\end{eqnarray*}
$$
が成立するとき、
$$
 M = USV^{T}
$$
と特異値分解できる。
特異値は$M^TM$もしくは$MM^T$の固有値の正の平方根。
$U$は$MM^T$の固有値の大きい順に対応する固有ベクトルを列成分に並べた行列。$V$は$M^TM$の固有値の大きい順に対応する固有ベクトルを列成分に並べた行列。
このときの固有ベクトルは正規化されている必要がある。
また、行列Sは特異値を対角成分にもつ行列（必ずしも正方行列ではない）
このとき、$u$を左特異ベクトル、$v$を右特異ベクトルという。

Mが$m \times n$なので、$\boldsymbol{v}$は$n \times 1$、$\boldsymbol{u}$は$m \times 1$となる。
よって、$U$は$m \times m$, $V$は$n \times n$となるので、$S$は$m \times n$行列になる。

$MM^T$の固有ベクトル$\boldsymbol{u}$を左特異ベクトル、$M^TM$の固有ベクトルを右特異ベクトルという。

- 例
$$
 M = 
 \left(
 \begin{array}{ccc}
 1 & 2 & 3 \\
 3 & 2 & 1
 \end{array}
 \right)
$$

$M M^T$を固有値分解する
$$
 MM^T = 
 \left(
 \begin{array}{ccc}

 3 & 2 & 1
 \end{array}
 \right)

 \begin{array}{cc}

 3 & 1
 \end{array}
 \right)

 \left(
 \begin{array}{cc}

 \end{array}
 \right)
$$

より、
$$
 \begin{array}{|cc|}
 14 - \lambda & 10 \\
 10 & 14 - \lambda
 \end{array}
 = 0 
$$

$$
 \lambda = 24,\  4
$$

- $\lambda=24$
のとき、固有ベクトルは
$$

$$

- $\lambda=4$のとき、固有ベクトルは

 \boldsymbol{x} =

 \end{array}
 \right)
$$

以上より、
$$
 \left(

 \end{array}
 \right)

 \end{array}

$$

次に、
